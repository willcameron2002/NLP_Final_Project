```python
import pandas as pd
```

```python
def Convert2Numbers(data):
    new_rt = []
    for rt in data['Retweets']:
        if type(rt) == int:
            new_rt.append(rt)
        elif rt[-1] == 'K':
            new_rt.append(int(float(rt[:-1]) * 1000))
        else:
            new_rt.append(int(rt))
    new_l = []
    for l in data['Likes']:
        if type(l) == int:
            new_l.append(l)
        elif l[-1] == 'K':
            new_l.append(int(float(l[:-1]) * 1000))
        else:
            new_l.append(int(l))
    
    data['Retweets'] = new_rt
    data['Likes'] = new_l
    data['Interaction Count'] = data['Retweets'] + data['Likes']
    
    ints = []
    for item in data['Interaction Count']:
        if item < 10:
            ints.append(item + 10)
        else:
            ints.append(item)
        
    data['Interaction Count'] = ints
    
    return
```
